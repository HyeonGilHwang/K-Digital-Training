```python
# select 문법으로 해보기

# BeautifulSoup로 크롤링 기본
import requests
import bs4

req = requests.get("https://www.hankookilbo.com/")

# print(req)
# <Response [200]>

hankook_html = req.text
bs_obj = bs4.BeautifulSoup(hankook_html, "lxml")

ul = bs_obj.select("div.content > h2") 

#print(ul[0].text)

for title in ul:
    title_1 = title.text
    print(title_1)
```

```python
ul1 = bs_obj.select("div.content > p") 

# print(len(ul1))

for aaa in ul1:
    aaa_1 = aaa.text
    print(aaa_1)
```

```python
import requests
import bs4

req = requests.get("https://tv.naver.com/r/")

# print(req)
# <Response [200]>

naver_tv_html = req.text
bs_obj = bs4.BeautifulSoup(naver_tv_html, "lxml")

ul = bs_obj.select("dt.title > a") 

no = 0

for i in ul:
    no += 1
    print(f'현재 인기 영상 {no}위는', i.text.replace("\n", "") , '입니다.')
    if no == 20:
        break
```
