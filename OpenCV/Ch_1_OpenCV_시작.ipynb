```python
img = cv2.imread('cat.bmp', cv2.IMREAD_GRAYSCALE)
# cv2.IMREAD_COLOR: BGR color로 읽기 --> default
# cv2.IMREAD_GRAYSCALE: 그레이 color로 읽기
# cv2.IMREAD_UNCHAGED: 파일 속성대로 읽기
# retval: numpy.ndarray로 반환

# print(img.shape)

# img 업로드 실패했을 경우 경고 문자!!
if img is None:
    print('Image load failed')
    sys.exit()
    
# cv2.namedWindow('Hello OpenCV') # 창만 뜸, 창크기 조절
cv2.imshow('Hello OpenCV', img) # Hello OpenCV 라는 새창을 열어라

# Keyboard 입력 (모든 키) 을 기다리는 동안 영상을 출력 
cv2.waitKey() 

# 키보드 입력을 받으면 화면에 나타난 모든 창을 닫음
# 아무키나 클릭하면 
cv2.destroyAllWindows() 
```

### image 저장하기: cv2.imwrite()

```python
# cv2.imwrite(filename, img, params =None) -> retval

# filename: 저장할 파일 이름
# img: 저장할 자료 (numpy.ndarray)
# params: 파일저장 옵션
# retval: 성공 True, 실패 False
```

### 새창띄우기 : cv2.namedWindow()

```python
# cv2.namedWindow(winname, flags =None) -> None

# winname: 창 이름
# flags:
# cv2.WINDOW_NORMAL: 영상크기를 창 크기에 맞게 지정
# cv2.WINDOW_AUTOSIZE: 창크기를 영상 크기에 맞게 변경
```

### image 저장하기: cv2.imshow()

```python
# imshow(winname, mat) -> None?

# winname: 창 이름
# mat: numpy.ndarray (default = uint8)
# uint16, int32의 경우 255로 나누어서 출력
# float32, float64의 경우 0 ~1로 만든 후 행렬값에 255를 곱해서 출력
```

### 키보드 입력대기: cv2.waitKey()

```python
# waitKey([, delay]) -> retval
# delay: Delay in milliseconds. 0 is the special value that means "forever"

# retval
```

```python
# img = cv2.imread('cat.bmp')
imgBGR = cv2.imread('cat.bmp', cv2.IMREAD_COLOR)
# img = cv2.imread('cat.bmp', cv2.IMREAD_UNCHANGED)


if img is None:
    print('Image load failed')
    sys.exit()

# BGR to RGB
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

## plt를 이용한 영상출력
plt.imshow(imgRGB)
plt.axis('off')
plt.show()
# plt로 이미지를 부를때는 cvtColor로 rgb 순서를 변환해야함
```

```python
## gray scale 영상으로 출력
imgGray = cv2.imread('cat.bmp', cv2.IMREAD_GRAYSCALE)

## 두 개의 영상을 모두 출력
plt.subplot(131), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(132), plt.imshow(imgGray, cmap = 'gray'), plt.axis('off') # 흑백으로 나타낼때 cmap = ' gray'
plt.subplot(133), plt.imshow(imgGray), plt.axis('off')
plt.show()
```

## 이미지 슬라이드쇼

```python

import sys
# Wild card
import glob
import cv2

img_files = glob.glob('./images/*.jpg')

for f in img_files:
    print(f)
    
# 전체 화면으로 'image' 창 생성
cv2.namedWindow('image', cv2.WINDOW_NORMAL) 
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN) 
# 전체 화면 설정 -> 항상 nameWindow 는 WINDOW_NORMAL로 해야함

cnt = len(img_files)
idx = 0

while True:
    img = cv2.imread(img_files[idx])
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(1000) == 27: #ESC
        break
    
    idx += 1
    if idx >= cnt:
        idx = 0
        
cv2.destroyAllWindows()
```
