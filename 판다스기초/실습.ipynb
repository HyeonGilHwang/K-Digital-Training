```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]
```

```python
df.head().sum()
```

```python
def missing_value(x):
    return x.isnull()

def missing_count(x):
    return missing_value(x).sum()

def totoal_number_missing(x):
    return missing_count(x).sum()
```

```python
result_value = df.pipe(totoal_number_missing)
result_value
```

```python
import pandas as pd
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
print(df1, '\n\n')
print(df2)
```

```python
result1 = pd.concat([df1, df2], ignore_index =True) 
# 위아래 붙인다 sas에서 set이랑 같은 의미
# ignore_index =True 기존 인덱스 무시하고 새로 리셋
result1
```

```python
result1 = pd.concat([df1, df2],axis =1)  
# axis =1   -> 인덱스 기준으로 옆으로 붙인다
result1
```

```python
result1 = pd.concat([df1, df2],axis =1, join = 'inner')  
# join = 'inner' -> 두 개의 데이터 모두 있는 것만
result1
```

```python
result1 = pd.concat([df1, df2],axis =1, join = 'outer')  
# join = 'outer' -> 모두 출력 default
result1
```
