```python
'''
concat으로 데이터프레임을 합쳐보자.
보통 데이터는 완성된 raw 데이터만 다루기 때문에 좌우로만 합친다.

'''
import pandas as pd
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
print(df1, '\n\n')
print(df2)
```

```python
'''
기본적으로 없는 부분은 NaN값이 채워짐을 이해하자.
join 옵션이 없다고 생각하고, 먼저 암기하자.
'''
# 2개의 데이터프레임을 위 아래 행 방향으로 연결하기 
result1 = pd.concat([df1, df2])
print(result1, '\n')


# ignore_index=True 옵션 이해하기
result2 = pd.concat([df1, df2], ignore_index=True)
print(result2, '\n')

# 2개의 데이터프레임을 좌우 열 방향으로 연결하기 
result3 = pd.concat([df1, df2], axis=1)
print(result3, '\n')

# join='inner' 옵션 적용하기(교집합)
result3_in = pd.concat([df1, df2], axis=1, join='inner')
print(result3_in, '\n')

# join='outer'는 디폴트 값이다.
result3_in = pd.concat([df1, df2], axis=1, join='outer')
print(result3_in, '\n')
```

```python
'''
series 포함한 concat 확인하기
사실 df에 series를 concat으로 추가하는 방식은 잘 안쓴다.
df['새로운 열이름'] = ['값1', '값2', ...]   이런식으로 주로 쓴다.
'''
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name='e')
sr2 = pd.Series(['f0', 'f1', 'f2'], name='f', index=[3, 4, 5])
sr3 = pd.Series(['g0', 'g1', 'g2', 'g3'], name='g')
print(sr1, '\n')
print(sr2, '\n')
print(sr3, '\n')
```

```python
'''
결과값은 참고만 하자
'''
result4 = pd.concat([df1, sr1], axis=1)
print(result4, '\n')

# df2과 sr2을 좌우 열 방향으로 연결하기
result5 = pd.concat([df1, sr2], axis=1, sort=True)
print(result5, '\n')

# sr1과 sr3을 좌우 열 방향으로 연결하기
result6 = pd.concat([sr1, sr3], axis=1)
print(result6, '\n')

# 위아래로 연결하기
result7 = pd.concat([sr1, sr3], axis=0)
print(result7, '\n')
```

```python
'''
두 데이터프레임이 같은 컬럼 이름인 경우의 concat
'''
import pandas as pd
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'number'])
df2 = pd.DataFrame([['c', 3], ['d', 4]],
                   columns=['letter', 'number'])

result = pd.concat([df1, df2], ignore_index = True)
print(result)
```

```python
'''
두 데이터프레임이 다른 컬럼 이름인 경우의 concat
'''
import pandas as pd

df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'number'])
df2 = pd.DataFrame([['c', 3], ['d', 4]],
                   columns=['letter', 'not_number'])
result = pd.concat([df1, df2], ignore_index=True)
print(result)

```

```python
'''
sort : 컬럼을 이름순으로 출력
inner : 합집합
inner 포함이 변화된 부분을 확인
'''
import pandas as pd
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'int'])
df3 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']],
                   columns=['letter', 'int', 'animal'])
result = pd.concat([df1, df3], join="inner", sort=True, ignore_index=True)
print(result)
```

```python
'''
join='outer'인 경우 확인
'''
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['a', 'b'])
df3 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']],
                   columns=['e', 'c', 'b'])
result = pd.concat([df1, df3], join="outer", sort=True, ignore_index=True)
print(result)
```

```python
'''
merge()
디폴트 : pd.merge(df_left, df_right, how='inner', on=None)
- on=None이므로 두 데이터의 공통 열이름을 기준으로 inner(교집합) 조인
- how : 
  - right, left : 선택한 df의 모든 키값을 포함하여 합한다.
  - outer : 한쪽만 있어도 보여준다
  - inner : 둘다 있어야 보여준다.
  - on='열' : 열을 키로 병합한다.
'''
dummy_data1 = { 
    'id': ['1', '2', '3', '4', '5'], 
    'Feature1': ['A', 'C', 'E', 'G', 'I'], 
    'Feature2': ['B', 'D', 'F', 'H', 'J']}
df1 = pd.DataFrame(dummy_data1, columns = ['id', 'Feature1', 'Feature2'])
print(df1, '\n')

dummy_data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['A', 'M', 'O', 'Q', 'S'],
        'Feature2': ['B', 'N', 'E', 'R', 'T']}
df2 = pd.DataFrame(dummy_data2, columns = ['id', 'Feature1', 'Feature2'])
print(df2, '\n')

dummy_data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Feature3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}
df3 = pd.DataFrame(dummy_data3, columns = ['id', 'Feature3'])
print(df3, '\n')
```

```python
'''
옵션 없이 merge 한경우에 겹치는 행만 출력 됨을 확인
'''
pd.merge(df1, df2)
```

```python
'''
id 열을 기준으로 merge한 경우
왼쪽 열의 컬럼에 _x가 붙고
오른쪽 열의 컬럼에 _y가 붙음을 확인
'''
pd.merge(df1, df2, on='id')
```

```python
'''
how='outer'를 추가한 경우 결과값을 확인
'''
pd.merge(df1, df2, on='id', how='outer')
```

```python
'''
how='letf'를 추가한 경우 결과값을 확인
'''
pd.merge(df1, df2, on='id', how='left')
```

```python
'''
how='right'를 추가한 경우 결과값을 확인
'''
pd.merge(df1, df2, on='id', how='right')
```

```python
'''
위의 예에서는 각각의 열에 on을 활용해 공통으로 기준열을 잡았지만, df마다 다른 열을 기준으로 잡기위해서는 아래와 같이 사용함을 이해
'''
'''
df1의 Feature1의 값을 기준으로 각각의 값을 df2의 Feature2에 같은 값이 있으면, 보여진다. 
아래는 'E'값 하나가 같기 때문에 하나의 열이 나타는 것이다
'''
pd.merge(df1, df2, how='left', left_on='Feature1', right_on='Feature2')
```

```python
'''


groupby


'''
import numpy as np
df = pd.DataFrame(
    {
        'A' : ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
        'B' : ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
        'C' : np.random.randn(8),
        'D' : np.random.randn(8)
    })
df
```

```python
'''
아래의 형태로 많이 쓰인다.
A열을 기준으로 각각의 범주화된 값들끼리의 합을 구하여 보여준다.

예시의 sum() 함수 대신 mean(), max(), min(), count(), size(), var(), std(), describe() 등의 내장함수가 사용가능함을 이해하자
'''
df.groupby('A').max()
```

```python
'''
아래와 같은 방식으로 사용가능함을 이해하자.

groupby된 A의 원소를 기준으로 B의 원소를 groupby한다고 이해하자.
'''
df.groupby(['A','B']).sum()
```

```python
'''
이러한 형식으로 시리즈 A의 범주화된 값들을 따로 뽑을 수 있음을 이해하자
'''
grouped = df.groupby(['A'])
for key, group in grouped:
    print(key)
    print(group)
```

```python
'''
A 내부부의 범주화된 값들의 데이터만 get_group를 활용하여 뽑을 수 있음을 이해하자
'''
grouped.get_group('foo')
```

```python
'''
agg를 통해 apply처럼 함수를 적용할 수 있다.
아래에 포함된 max는 내장함수라고 생각하자.
min, mean 등등이 들어가도 된다.
이때 grouped.agg(max)은 grouped.agg('max')과 같다.
'''
grouped = df.groupby(['A'])
grouped.agg('mean')
```

```python
'''
여러개의 내장함수가 포함되면 리스트에 담아줌을 이해하자.
이때 max왜 min은 따옴표가 없어도 되지만, mean이나 std는 따옴표 안에 넣어줘야 에러없이 실행됨을 이해하자
'''
grouped = df.groupby(['A'])
grouped.agg([max, min, 'mean', 'std'])
```
