```python
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]
ndf.head()
```

```python
ndf.plot(kind='scatter', x='weight', y='mpg',  c='coral', s=10, figsize=(10, 5))
plt.show()
```

```python
X = ndf[['weight']]
y = ndf['mpg']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.3,
                                                   random_state = 10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# sklearn 라이브러리에서 선형회귀분석 모듈 가져오기
from sklearn.linear_model import LinearRegression
```

```python
# train data를 가지고 모형 학습
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 학습을 마친 모형에 test data를 적용하여 결정계수(R-제곱) 계산
r_square = lr.score(X_test, y_test)
print(r_square)
```

```python
# 회귀식의 기울기
print('기울기 a: ', lr.coef_)
```

```python
# 회귀식의 y절편
print('y절편 b:', lr.intercept_)
```

```python
# 모형에 전체 X 데이터를 입력하여 예측한 값 y_hat을 실제 값 y와 비교 
y_hat = lr.predict(X)
```

```python
plt.figure(figsize=(10, 6))
ax1 = sns.distplot(y, hist=False, label="y")# 파란색이 실데이터
ax2 = sns.distplot(y_hat, hist=False, label="y_hat", ax=ax1)# 파란색이 예측데이터
plt.show() 
```

#### polynomial_regression

```python
from sklearn.preprocessing import PolynomialFeatures # 다항식 변환
```

```python
poly = PolynomialFeatures(degree = 2) # 2차항으로 만들겠다
X_train_poly = poly.fit_transform(X_train)
```

```python
X_train.shape
```

```python
X_train_poly.shape
```

```python
X_train_poly # 인터셉트, x, x^2
```

```python
# train data 학습
# trest data 평가, 검증

# 1. train data로 학습
plr = LinearRegression()   
plr.fit(X_train_poly, y_train)
```

```python
X_test_poly = poly.fit_transform(X_test)
r_square = plr.score(X_test_poly, y_test)
r_square
```

```python
y_hat_test = plr.predict(X_test_poly)
```

```python
fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(X_train, y_train, 'o', label = 'Train')
ax.plot(X_test, y_hat_test, 'r+', label = 'Predicted') # r+ 빨간색 플러스
ax.legend(loc = 'best')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
X_poly = poly.fit_transform(X)
y_hat = plr.predict(X_poly)
```

```python
plt.figure(figsize=(10, 6))
ax1 = sns.distplot(y, hist=False, label="y")
ax2 = sns.distplot(y_hat, hist=False, label="y_hat", ax=ax1)
plt.show()
plt.close()
```

#### 다중회귀분석

```python
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]

X = ndf[['cylinders','horsepower', 'weight']] # x1, x2, x3
y = ndf['mpg']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.3, random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
r_square = lr.score(X_test, y_test)
r_square
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_hat = lr.predict(X_test)
```

```python
plt.figure(figsize= (10,6))
ax1 = sns.distplot(y_test, hist=False, label="y_test")
ax2 = sns.distplot(y_hat, hist=False, label="y_hat", ax=ax1)
plt.show()
plt.close()
```
