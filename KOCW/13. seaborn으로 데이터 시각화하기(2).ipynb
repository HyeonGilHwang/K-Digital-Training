```python
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.pairplot(tips, hue = 'size', height = 4)
```

```python
sns.pairplot(tips, hue = 'day', height = 3)
```

```python
sns.barplot(x = 'day', y = 'tip', data = tips, hue = 'smoker')
```

```python
tips['day']. value_counts()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'size',hue = 'smoker', col_wrap =3) # col_wrap =3 한줄에 3개  
```

```python
pivot = tips.pivot_table(index = 'day', columns = 'smoker', values = 'tip')
```

```python
pivot
```

```python
sns.heatmap(pivot, annot = True, cmap = 'Blues')
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', size = 'smoker', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day')
```

```python
sns.catplot(x = 'smoker', y = 'tip', data= tips) # 카테코리컬 데이터 플랏
```
