```python
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(data=iris, hue = 'species', vars = ['petal_length', 'petal_width'], height = 4) # hue 종별로 색깔, height = 4 그래프 크기조절
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'scatter')
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'reg')
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'kde')
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'hex')
```

```python
sns.lmplot(x = 'petal_length', y = 'petal_width', hue = 'species', data= iris)
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(), annot = True)
```

```python
sns.heatmap(iris.corr(), annot = True, cmap = 'Blues')
```

```python
sns.heatmap(iris.corr(), annot = True, cmap = 'Reds')
```

```python
sns.heatmap(iris.corr(), annot = True, cmap = 'Greens')
```

```python
sns.scatterplot(x = 'petal_length', y = 'petal_width', data= iris)
```

```python
sns.scatterplot(x = 'petal_length', y = 'petal_width', size = 'species',data= iris)
```

```python
sns.relplot(x = 'petal_length', y = 'petal_width', hue = 'species', data= iris)
```
